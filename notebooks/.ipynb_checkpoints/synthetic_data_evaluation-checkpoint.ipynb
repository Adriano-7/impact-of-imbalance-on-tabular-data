# Evaluation of Synthetic Data Quality vs. Imbalance

## 1. Executive Summary

This notebook evaluates and compares the quality of synthetic data generated by **CTGAN** and **TVAE** across six imbalance ratios (IRs). The analysis is structured around three core quality dimensions: **Statistical Fidelity**, **Downstream Utility**, and **Privacy Risk**.

### Key Findings:
* **TVAE is the superior generator for this task**, especially under severe class imbalance (IR > 20:1). It consistently maintains high utility (PR-AUC > 0.63) and privacy (Authenticity > 0.23) where CTGAN's performance collapses.
* **CTGAN excels slightly in Fidelity (realism)**, achieving higher Precision and Density scores. However, this comes at a steep cost to privacy, with Authenticity scores dropping near zero at high IRs, suggesting it may be overfitting to the few real minority samples.
* **Diversity metrics (`ImprovedRecall`, `Coverage`) were uninformative**, with both generators achieving perfect scores across all conditions. This suggests they are not sensitive enough for this particular evaluation and have been excluded from the main visualizations for clarity.

**Recommendation:** For downstream classification tasks on this dataset, **TVAE is the recommended generator**. It strikes an excellent balance between generating useful, novel data without simply memorizing the training set.

## 2. Setup and Configuration

We'll start by importing libraries and defining key parameters for the analysis. A professional, colorblind-friendly palette is chosen for all visualizations to ensure clarity and accessibility.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from tqdm.notebook import tqdm
import warnings

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve, auc
from sklearn.preprocessing import MinMaxScaler

from pymdma.tabular.measures.synthesis_val import (
    ImprovedPrecision,
    ImprovedRecall,
    Density,
    Coverage,
    Authenticity,
)

warnings.filterwarnings("ignore")
```

```python
#   Configuration  
BASE_PATH = Path("../")
PROCESSED_PATH = BASE_PATH / "data" / "processed"
SYNTHETIC_PATH = BASE_PATH / "data" / "synthetic"
FIGURES_PATH = BASE_PATH / "results" / "figures"
FIGURES_PATH.mkdir(exist_ok=True, parents=True)

TARGET_FEATURE = "Severity"
IMBALANCE_RATIOS = [1, 5, 10, 20, 50, 100]
GENERATORS = ["ctgan", "tvae"]

#   Plotting Style  
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = 'sans-serif'

# Define a professional, colorblind-friendly color palette
palette = sns.color_palette("colorblind", n_colors=len(GENERATORS))
GENERATOR_STYLES = {
    "CTGAN": {"color": palette[0], "marker": "o", "linestyle": "-"},
    "TVAE": {"color": palette[1], "marker": "s", "linestyle": "--"},
}
```

## 3. Metric Calculation Pipeline

These helper functions encapsulate the logic for calculating utility and `pyMDMA` quality metrics. Adding docstrings improves readability and makes the code easier to maintain or reuse.

```python
def calculate_utility_metrics(train_df: pd.DataFrame, test_df: pd.DataFrame) -> dict:
    """
    Trains a RandomForest on synthetic data and evaluates its PR-AUC on real test data.

    Args:
        train_df (pd.DataFrame): The training dataset (synthetic).
        test_df (pd.DataFrame): The testing dataset (real).

    Returns:
        dict: A dictionary containing the PR-AUC score.
    """
    X_train = train_df.drop(columns=[TARGET_FEATURE])
    y_train = train_df[TARGET_FEATURE]
    X_test = test_df.drop(columns=[TARGET_FEATURE])
    y_test = test_df[TARGET_FEATURE]

    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)

    y_prob = model.predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, y_prob)
    
    return {"pr_auc": auc(recall, precision)}


def calculate_pyMDMA_metrics(real_df: pd.DataFrame, synthetic_df: pd.DataFrame) -> dict:
    """
    Calculates a suite of statistical quality and privacy metrics using pyMDMA.

    Args:
        real_df (pd.DataFrame): The original, real dataset.
        synthetic_df (pd.DataFrame): The generated synthetic dataset.

    Returns:
        dict: A dictionary of calculated scores.
    """
    real_features = real_df.drop(columns=[TARGET_FEATURE]).to_numpy()
    synthetic_features = synthetic_df.drop(columns=[TARGET_FEATURE]).to_numpy()

    metrics = {
        "precision": ImprovedPrecision(),
        "recall": ImprovedRecall(),
        "density": Density(),
        "coverage": Coverage(),
        "authenticity": Authenticity(),
    }
    
    results = {
        name: metric.compute(real_features, synthetic_features).dataset_level.value
        for name, metric in metrics.items()
    }
    return results
```

```python
#   Data Loading and Processing  
all_results = []
test_df = pd.read_csv(PROCESSED_PATH / "test.csv")

# Use tqdm for a clean progress bar
pbar = tqdm(total=len(IMBALANCE_RATIOS) * len(GENERATORS), desc="ðŸš€ Processing Experiments")

for ir in IMBALANCE_RATIOS:
    for generator in GENERATORS:
        real_path = PROCESSED_PATH / f"train_imbalanced_ir_{ir}.csv"
        synth_path = SYNTHETIC_PATH / f"train_imbalanced_ir_{ir}_balanced_by_{generator}_run1.csv"

        if not synth_path.exists():
            warnings.warn(f"File not found, skipping: {synth_path.name}")
            pbar.update(1)
            continue

        real_df = pd.read_csv(real_path)
        synthetic_df = pd.read_csv(synth_path)
        
        # Calculate all metrics
        pyMDMA_metrics = calculate_pyMDMA_metrics(real_df, synthetic_df)
        utility_metrics = calculate_utility_metrics(synthetic_df, test_df)

        # Store results
        result_entry = {
            "ir": ir,
            "generator": generator.upper(),
            **pyMDMA_metrics,
            **utility_metrics
        }
        all_results.append(result_entry)
        pbar.update(1)

pbar.close()
results_df = pd.DataFrame(all_results)
print("âœ… Evaluation complete.")
results_df.head()
```

## 4. Analysis and Visualization

### 4.1. At-a-Glance Performance Summary

Before diving into plots, a styled table provides a quick, high-level overview. We can immediately see which generator performs best for each metric at a given imbalance ratio.

**Observation**: **TVAE** consistently wins on **Utility (pr_auc)** and **Privacy (authenticity)**, while **CTGAN** has a slight edge in **Fidelity (precision, density)**.

```python
# Pivot the data for a summary view and apply styling
summary_pivot = results_df.pivot_table(
    index='ir', 
    columns='generator', 
    values=['precision', 'density', 'pr_auc', 'authenticity']
)

# Improve column names for clarity
summary_pivot.columns = [f'{col[1]}_{col[0]}' for col in summary_pivot.columns]
summary_pivot = summary_pivot.rename(columns=lambda c: c.replace('_auc', '-AUC'))


# Highlight the maximum value in each metric group across generators
styled_summary = summary_pivot.style.highlight_max(
    subset=[col for col in summary_pivot.columns if 'CTGAN' in col or 'TVAE' in col],
    props='font-weight:bold; background-color:#a2d2ff;', 
    axis=1
).format("{:.3f}")

display(styled_summary)
```

### 4.2. Performance Trends vs. Imbalance Ratio

The original 3x2 plot layout was cluttered. A more focused 2x2 grid on the **most informative metrics** tells a clearer story. We create a helper function `plot_metric_trend` to keep the code clean and reusable (DRY principle).

**Analysis of Trends:**
* **Fidelity & Density**: CTGAN consistently produces slightly more realistic data points. However, both models degrade as the imbalance ratio increases, since they have fewer minority examples to learn from.
* **Utility (PR-AUC)**: TVAE's utility remains remarkably stable and high, even at a 100:1 imbalance. CTGAN's utility plummets after the 20:1 mark, rendering it ineffective for the downstream task.
* **Privacy (Authenticity)**: This is the most dramatic difference. TVAE generates novel samples, maintaining a good authenticity score. CTGAN's authenticity score collapses, indicating it is likely memorizing and regenerating samples from the tiny minority class (a form of overfitting).

```python
def plot_metric_trend(data, metric, ax, title, y_label, y_lim=(0, None)):
    """A modular function to plot metric trends vs. imbalance ratio."""
    sns.lineplot(
        data=data, 
        x='ir', 
        y=metric, 
        hue='generator', 
        style='generator', 
        ax=ax,
        palette=[GENERATOR_STYLES[gen]['color'] for gen in data['generator'].unique()],
        markers=[GENERATOR_STYLES[gen]['marker'] for gen in data['generator'].unique()],
        dashes=[(1,0) if ls['linestyle']=='-' else (4,2) for gen, ls in GENERATOR_STYLES.items()],
        markersize=8, 
        linewidth=2.5
    )
    
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_ylabel(y_label, fontsize=12)
    ax.set_xlabel('Imbalance Ratio (Log Scale)', fontsize=12)
    ax.set_xscale('log')
    ax.set_xticks(IMBALANCE_RATIOS)
    ax.set_xticklabels([f"{ir}:1" for ir in IMBALANCE_RATIOS])
    ax.set_ylim(y_lim)
    ax.legend(title='Generator', loc='best')
    ax.grid(True, which="both", ls="--", c='0.85')

# Create a more focused 2x2 plot
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Key Quality Metrics vs. Imbalance Ratio', fontsize=20, fontweight='bold')

plot_metric_trend(results_df, 'precision', axes[0, 0], 'Fidelity: Improved Precision', 'Precision Score')
plot_metric_trend(results_df, 'density', axes[0, 1], 'Robust Fidelity: Density', 'Density Score')
plot_metric_trend(results_df, 'pr_auc', axes[1, 0], 'Utility: Downstream PR-AUC', 'PR-AUC Score', y_lim=(0, 1))
plot_metric_trend(results_df, 'authenticity', axes[1, 1], 'Privacy: Authenticity', 'Authenticity Score', y_lim=(0, 0.35))

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig(FIGURES_PATH / 'improved_quality_trends.png', dpi=300)
plt.show()
```

### 4.3. Multi-Dimensional Quality Profile (Radar Chart)

A radar chart is a powerful tool for comparing multiple metrics simultaneously. It helps visualize the **overall quality profile** of each generator and highlights their strengths and weaknesses in a single view.

Here, we normalize all metrics to a 0-1 scale to plot them on the same axes.

**How to Read the Charts:** A larger, more balanced shape indicates a better-performing generator across all dimensions.
* At **IR 5:1**, both generators have similar profiles.
* At **IR 20:1**, TVAE's shape is larger, showing its superiority in Utility and Privacy.
* At **IR 100:1**, the difference is stark. TVAE maintains a robust, well-rounded profile, while CTGAN's profile has collapsed towards the Fidelity axis, showing it is a "one-trick pony" that fails on other crucial dimensions.

```python
# Select and scale metrics for the radar chart
metrics_for_radar = ['precision', 'density', 'pr_auc', 'authenticity']
radar_df = results_df.copy()

# Scale metrics to a common [0, 1] range for fair comparison on the radar plot
scaler = MinMaxScaler()
radar_df[metrics_for_radar] = scaler.fit_transform(radar_df[metrics_for_radar])

# Define the radar chart plotting function
def plot_radar_chart(data, ir_level, ax):
    """Generates a single radar chart for a given imbalance ratio."""
    df_ir = data[data['ir'] == ir_level]
    labels = metrics_for_radar
    num_vars = len(labels)
    
    # Create angles for the radar chart
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1] # Close the loop

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    for gen, style in GENERATOR_STYLES.items():
        values = df_ir[df_ir['generator'] == gen][labels].values.flatten().tolist()
        values += values[:1] # Close the loop
        ax.plot(angles, values, color=style['color'], linewidth=2, linestyle=style['linestyle'], label=gen)
        ax.fill(angles, values, color=style['color'], alpha=0.25)

    ax.set_rlabel_position(180 / num_vars)
    ax.set_xticks(angles[:-1], [l.replace('_', ' ').title() for l in labels], size=11)
    ax.set_yticks([0.2, 0.4, 0.6, 0.8])
    ax.set_yticklabels(["0.2", "0.4", "0.6", "0.8"], color="grey", size=9)
    ax.set_ylim(0, 1)
    ax.set_title(f"IR = {ir_level}:1", size=14, weight='bold', position=(0.5, 1.1))

#   Create Radar Plots  
fig, axes = plt.subplots(figsize=(18, 5), ncols=3, subplot_kw=dict(projection='polar'))
fig.suptitle('Generator Quality Profile at Different Imbalance Ratios', fontsize=20, fontweight='bold', y=1.05)

# Plot for specific IR levels to show the change
plot_radar_chart(radar_df, 5, axes[0])
plot_radar_chart(radar_df, 20, axes[1])
plot_radar_chart(radar_df, 100, axes[2])

# Add a single legend for the entire figure
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', fontsize=12)

plt.tight_layout()
plt.savefig(FIGURES_PATH / 'improved_radar_profiles.png', dpi=300, bbox_inches='tight')
plt.show()
```
